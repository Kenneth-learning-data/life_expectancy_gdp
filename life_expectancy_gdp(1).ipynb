```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from decimal import Decimal
```

```python
df = pd.read_csv("all_data.csv")
```

```python
df.columns = ['country', 'year', 'life expectancy','gdp']
```

```python

print(df.head(10))
print(df.describe())
print(df.info())
```

```python
#total 4 columns : country/ year /life expectancy / gdp

print(df.country.nunique())
#6 countries

country = df.country.unique()
print(country)
#['Chile' 'China' 'Germany' 'Mexico' 'United States of America' 'Zimbabwe']

print(df.year.unique())
#16
#2000 -2015
```

```python
def six_countries_LE(country):
    country_df= df[df.country== country]
    plt.scatter(country_df.year, country_df['life expectancy'])
    plt.legend(['Chile', 'China' ,'Germany' ,'Mexico', 'United States of America', 'Zimbabwe'], loc="best")
    plt.xlabel("year")
    plt.ylabel('Life expectancy')
    plt.title(country +" life expectancy over years")
    plt.show()
  

six_countries_LE("Chile")
six_countries_LE("China")
six_countries_LE("Mexico")
six_countries_LE("United States of America")
six_countries_LE("Zimbabwe")

#increasing trends in 6 countries 
#Except US, all countries have similar trends, increasing slightly
```

```python
# All countries have a increasing trend in Life expectancy over years
```

```python
def six_countries_gdp(country):
    country_df= df[df.country== country]
    plt.scatter(country_df.year, country_df['gdp'])
    plt.xlabel("year")
    plt.ylabel('GDP')
    plt.title(country +" GDP over years")
    plt.show()

six_countries_gdp("Chile")
six_countries_gdp("China")
six_countries_gdp("Mexico")
six_countries_gdp("United States of America")
six_countries_gdp("Zimbabwe")
```

```python
# All countries have a increasing trend in GDP over years
```

```python
# Is there a correlation between GDP and life expectancy of a country?
```

```python
def gdp_and_le(country):
    country_df= df[df.country== country]
    plt.scatter(country_df.gdp, country_df['life expectancy'])
    plt.xlabel("gdp")
    plt.ylabel('Life expectancy')
    plt.title(country+ "relationship betwenn gdp and life expectancy")
    plt.show()


gdp_and_le("Chile")
gdp_and_le("China")
gdp_and_le("Mexico")
gdp_and_le("United States of America")
gdp_and_le("Zimbabwe")
```

```python
# Higher GDP, higher Life expectancy
```
